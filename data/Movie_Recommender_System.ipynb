# Data Preprocessing

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.genres[0]
```

```python
import ast
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
convert(df.genres[0])
```

```python
df.genres = df.genres.apply(convert)
```

```python
df.keywords[0]
```

```python
df.keywords = df.keywords.apply(convert)
```

```python
df.head(2)
```

```python
df.cast[0]
```

```python
def convert3(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      l.append(i['name'])
      counter += 1
    else:
      break
  return l
```

```python
convert3(df.cast[0])
```

```python
df.cast = df.cast.apply(convert3)
```

```python
df.crew[0]
```

```python
def find_director(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
```

```python
df.crew = df.crew.apply(find_director)
```

```python
df.overview = df.overview.apply(lambda x:x.split())
```

```python
df.head(2)
```

```python
df.genres = df.genres.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.keywords = df.keywords.apply(lambda x:[i.replace(" ","") for i in x])
df.cast = df.cast.apply(lambda x:[i.replace(" ","") for i in x])
df.crew = df.crew.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head(2)
```

```python
df['tags'] = df.overview + df.genres + df.keywords + df.cast + df.crew
```

```python
films = df[['movie_id','title','tags']]
films.tags = films.tags.apply(lambda x:" ".join(x))
films.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
films.tags = films.tags.apply(stem)
```

```python
films.tags[0]
```

```python
films.tags= films.tags.apply(lambda x:x.lower())
```

```python
# Test vectors
# Bag of words
#   -->> 1.concatenate all tags
#   -->> 2. Extract most frequent words

# Removing stop words : 'are','is','from','of' etc.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(films.tags).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sns.heatmap(pd.DataFrame(similarity))
#diagonal is always 1 because of same data point
```

```python
def recommend(movie):
  movie_index = films[films['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(films.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
